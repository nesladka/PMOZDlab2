```python
import matplotlib as plt
import pandas as pd
```

```python
data = pd.read_csv('metro.csv')
data
```

```python

grouped_data = data.groupby('name')['count_carriages'].sum()

plt.figure(figsize=(10, 8))
plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%')
plt.title('Количество действующих вагонов на ветках метро')
plt.axis('equal')  
plt.show()
```

```python
plt.barh(data['name'], data['lenth'], color='skyblue')
plt.xlabel('Протяженность')
plt.ylabel('Ветки метро')
plt.title('Протяженность веток метро')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(data['name'], data['count_station'], color='skyblue')
plt.xlabel('Ветки метро')
plt.ylabel('Количество станций')
plt.title('Количество станций на ветках метро')
plt.xticks(rotation=90)  
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
